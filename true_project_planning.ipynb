```r
library(tidyverse)

players <- read_csv("players.csv")
head(players)

sessions <- read_csv("sessions.csv")
head(sessions)
```

```r
players_summary <- summary(players)

players_summary

sessions_summary <- summary(sessions)

sessions_summary
```

```r
players_mean <- players |>
    select(played_hours, Age)|>
    colMeans(na.rm = TRUE)

players_mean

players_tibble <- players_mean |>
    as_tibble()

players_tibble
```

```r
age_time_plot <- players |>
    mutate(experience = as.factor(experience))|>
    select(Age, played_hours, experience)|>
    ggplot(aes(x = Age, y = played_hours, colour = experience))+
    geom_point()+
    labs(title = "Age vs Time Played", x = "Age (years)", y = "Time Played (hours)", colour = "experience level")

age_time_plot
```
